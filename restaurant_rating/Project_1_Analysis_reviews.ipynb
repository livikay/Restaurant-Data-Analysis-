```python
X = [3, 3.5, 4, 4.5, 5]
Yprereviews = [51, 78, 107, 57, 7]
Zpostreviews = [129, 201, 290, 202, 36]

X_axis = np.arange(len(X))
   
plt.bar(X_axis - 0.2, Yprereviews, 0.4, label = 'Precovid Ohio 2019')
plt.bar(X_axis + 0.2, Zpostreviews, 0.4, label = 'Postcovid Ohio 2020')

plt.xticks(X_axis, X)
plt.xlabel("Review Ratings")
plt.ylabel("Number of Reviews")
plt.title("Precovid Ohio 2019 vs Postcovid Ohio 2020")
plt.savefig("PrecovidOhio2019_vs_PostcovidOhio2020.png")
plt.legend()
plt.show()
```
